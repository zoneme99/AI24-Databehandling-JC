```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
#Load data
print("a and b)--------------")
college = pd.read_csv('College.csv')
college = college.rename({'Unnamed: 0': 'College'} , axis =1)
college = college.set_index('College')
college
```

```python
print("c)--------------")
college.describe()
```

```python
print("d)--------------")
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
print("e)--------------")
college.boxplot(column="Outstate", by="Private")
```

```python
print("f)--------------")
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
display(college)
print(college['Elite'].value_counts())
college.boxplot(column="Outstate", by="Elite")

```

```python
print("g)--------------")
fig, ((top10, accept), (enroll, top25)) = plt.subplots(2, 2)
fig.set_size_inches(12, 8)
top10.hist(college['Top10perc'])
top10.set_title('Top10perc')
accept.hist(college['Accept'])
accept.set_title('Accept')
enroll.hist(college['Enroll'])
enroll.set_title('Enroll')
top25.hist([college['Top25perc']])
top25.set_title('Top25perc')
plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize = (12,4))
ax[0].set_title('Expenditures for students')
ax[1].set_title('Personal expenses')
ax[2].set_title(' grad rate mean')
expends = college.groupby('Elite')['Expend'].mean()
expends.plot(kind='bar', ax=ax[0])
personal = college.groupby('Elite')['Personal'].mean()
personal.plot(kind='bar', ax=ax[1])
gradrate = college.groupby('Elite')['Grad.Rate'].mean()
gradrate.plot(kind='bar', ax=ax[2])
```
